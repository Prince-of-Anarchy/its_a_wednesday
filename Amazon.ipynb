```python
import pandas as pd

# Load the amazon.csv dataset into a pandas DataFrame
try:
    df = pd.read_csv('amazon.csv')
    print("Dataset 'amazon.csv' loaded successfully!")
    display(df.head())
except FileNotFoundError:
    print("Error: amazon.csv not found. Please make sure the file is uploaded.")
except Exception as e:
    print(f"Error loading the dataset: {e}")
```

```python
# Identify columns that likely contain numeric data but are of object type
cols_to_convert = ['discounted_price', 'actual_price', 'discount_percentage', 'rating']

for col in cols_to_convert:
    if col in df.columns:
        # Remove currency symbols and commas, then convert to numeric
        if col in ['discounted_price', 'actual_price']:
            df[col] = df[col].astype(str).str.replace('₹', '').str.replace(',', '')
        elif col == 'discount_percentage':
            df[col] = df[col].astype(str).str.replace('%', '')
        # Convert to numeric, coercing errors will turn unconvertible values into NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Display the data types after conversion
print("Data types after conversion:")
df.info()
```

**Reasoning**:
Calculate and print the number and percentage of missing values for each column in the DataFrame.



```python
# Calculate the number of missing values for each column
missing_values_count = df.isnull().sum()

# Calculate the percentage of missing values for each column
missing_values_percentage = (missing_values_count / len(df)) * 100

# Print the count of missing values per column
print("Number of missing values per column:")
print(missing_values_count)

# Print the percentage of missing values per column, formatted to two decimal places
print("\nPercentage of missing values per column:")
print(missing_values_percentage.round(2))
```

**Reasoning**:
Create a heatmap to visualize the distribution of missing values across the DataFrame.



```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()
```

**Reasoning**:
Fill the missing values in 'rating' and 'rating_count' columns with their respective median values and then verify that there are no missing values left.



```python
# Fill missing values in 'rating' with the median
df['rating'].fillna(df['rating'].median(), inplace=True)

# Fill missing values in 'rating_count' with the median
df['rating_count'].fillna(df['rating_count'].median(), inplace=True)

# Verify that there are no more missing values
print("Number of missing values after imputation:")
print(df.isnull().sum())
```

**Reasoning**:
Create a bar chart to visualize the percentage of missing values for each column.



```python
plt.figure(figsize=(12, 7))
missing_values_percentage.plot(kind='bar')
plt.title("Percentage of Missing Values per Column")
plt.xlabel("Columns")
plt.ylabel("Percentage of Missing Values")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Identify and explain the chosen pairs of variables for bivariate analysis based on the identified numerical and categorical columns.



```python
numerical_cols = ['discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']
categorical_cols = ['product_id', 'product_name', 'category', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']

# Chosen pairs for bivariate analysis:

# Numerical-Numerical Pair:
# 1. discounted_price vs. actual_price
#    Reasoning: To understand the relationship between the discounted price and the original price,
#    and to see if there's a general pattern in pricing strategies.

# Numerical-Categorical Pair:
# 2. rating vs. category
#    Reasoning: To investigate if there are differences in average ratings across different product categories.
#    This can help identify categories with generally higher or lower customer satisfaction.

# Categorical-Categorical Pair:
# 3. category vs. product_name (or product_id)
#    Reasoning: To explore the distribution of products within different categories and identify the most
#    frequent product names or IDs within top categories. This can reveal popular products or diverse categories.

print("Chosen Variable Pairs for Bivariate Analysis:")
print("1. Numerical-Numerical: discounted_price vs. actual_price")
print("2. Numerical-Categorical: rating vs. category")
print("3. Categorical-Categorical: category vs. product_name")
```

```python
# Calculate Pearson correlation coefficient
correlation = df['actual_price'].corr(df['discounted_price'], method='pearson')
print(f"Pearson correlation between Actual Price and Discounted Price: {correlation:.2f}")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='actual_price', y='discounted_price', data=df)
plt.title("Scatter Plot of Discounted Price vs. Actual Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")
plt.show()
```

**Reasoning**:
Calculate the mean rating for each category, print the sorted mean ratings, and create box and violin plots for the top N categories to visualize the distribution of ratings.



```python
# Calculate the mean rating for each category
mean_ratings_by_category = df.groupby('category')['rating'].mean()

# Print the mean ratings for each category, sorted in descending order
print("Mean Rating by Category (Descending Order):")
print(mean_ratings_by_category.sort_values(ascending=False))

# Determine the top N categories based on the number of products
n_top = 15
top_categories = df['category'].value_counts().nlargest(n_top).index.tolist()

# Filter the DataFrame to include only the top categories
df_top_categories = df[df['category'].isin(top_categories)]

# Create a box plot for the distribution of ratings in top categories
plt.figure(figsize=(15, 8))
sns.boxplot(x='rating', y='category', data=df_top_categories, order=top_categories)
plt.title(f'Distribution of Ratings for Top {n_top} Categories (Box Plot)')
plt.xlabel('Rating')
plt.ylabel('Category')
plt.tight_layout()
plt.show()

# Create a violin plot for the distribution of ratings in top categories
plt.figure(figsize=(15, 8))
sns.violinplot(x='rating', y='category', data=df_top_categories, order=top_categories)
plt.title(f'Distribution of Ratings for Top {n_top} Categories (Violin Plot)')
plt.xlabel('Rating')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Calculate and display the frequency counts and percentages for the 'product_name' column and then analyze the relationship between 'category' and 'product_name' by finding the most frequent 'product_name' within each category and finally determine the top N categories based on the number of products and filter the dataframe accordingly.



```python
# Calculate the frequency counts for 'product_name'
freq_counts = df['product_name'].value_counts()

# Calculate the frequency percentages for 'product_name'
freq_percentages = df['product_name'].value_counts(normalize=True) * 100

# Print the frequency counts for the top 10 most frequent product names
print("Top 10 Most Frequent Product Names (Counts):")
print(freq_counts.head(10))

# Print the frequency percentages for the top 10 most frequent product names
print("\nTop 10 Most Frequent Product Names (Percentages):")
print(freq_percentages.head(10).round(2))

# Analyze the relationship between 'category' and 'product_name' by finding the most frequent 'product_name' within each category
most_frequent_product_by_category = df.groupby('category')['product_name'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)

# Print the most frequent product name for each category
print("\nMost Frequent Product Name by Category:")
print(most_frequent_product_by_category)

# Determine the top N categories based on the number of products (using N=15 as done previously)
n_top = 15
top_categories = df['category'].value_counts().nlargest(n_top).index.tolist()

# Filter the DataFrame to include only the top categories
df_top_categories = df[df['category'].isin(top_categories)].copy()

# Display the first few rows of the filtered DataFrame
print(f"\nDataFrame filtered for top {n_top} categories:")
display(df_top_categories.head())
```

**Reasoning**:
Aggregate the data to count the occurrences of each product name within each of the top categories and then plot a stacked bar chart to visualize the distribution of the top product names within the top categories.



```python
# Aggregate the data to count the occurrences of each product name within each top category
category_product_counts = df_top_categories.groupby('category')['product_name'].value_counts().unstack(fill_value=0)

# Select the top M product names within the top N categories for visualization (choose a reasonable M, e.g., 5)
M_top_products = 5
top_products_in_top_categories = category_product_counts.apply(lambda x: x.nlargest(M_top_products), axis=1)

# Plot a stacked bar chart
plt.figure(figsize=(18, 10))
top_products_in_top_categories.plot(kind='bar', stacked=True, figsize=(18, 10))
plt.title(f'Distribution of Top {M_top_products} Product Names within Top {n_top} Categories')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

**Reasoning**:
The stacked bar plot is too cluttered with many product names. To improve readability and focus on the most prominent products within each category, it is better to select only the top M products *overall* from the entire dataset and then visualize their distribution across the top N categories.



```python
# Select the top M product names overall from the entire dataframe
M_top_products_overall = 10  # Choose a reasonable number for top products overall
top_product_names_overall = df['product_name'].value_counts().nlargest(M_top_products_overall).index.tolist()

# Filter the DataFrame to include only the top categories and the top product names overall
df_filtered_top_products = df_top_categories[df_top_categories['product_name'].isin(top_product_names_overall)].copy()

# Aggregate the data to count the occurrences of the selected top product names within each of the top categories
category_top_product_counts = df_filtered_top_products.groupby('category')['product_name'].value_counts().unstack(fill_value=0)

# Plot a stacked bar chart with the filtered data
plt.figure(figsize=(18, 10))
category_top_product_counts.plot(kind='bar', stacked=True, figsize=(18, 10))
plt.title(f'Distribution of Top {M_top_products_overall} Overall Product Names within Top {n_top} Categories')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Product Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
